```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
x
```

```python
y  = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
cosy = np.cos(x)
plt.plot(x,y)
plt.plot(x, cosy)
plt.axis([-1, 11, -2, 2])
plt.show()
```

Scatter Plot

```python
plt.scatter(x, y)
plt.scatter(x, cosy)
plt.show()
```

```python
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)
plt.scatter(x, y, alpha=0.5)
plt.show()
```

```python
from sklearn import datasets
iris = datasets.load_iris()
iris.keys()
iris.data.shape
iris.feature_names
iris.target.shape
```

```python
x = iris.data[:,:2]
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
y = iris.target
plt.scatter(x[y==0,0],x[y==0,1])
plt.scatter(x[y==1,0],x[y==1,1])
plt.scatter(x[y==2,0],x[y==2,1])
```

```python
x = iris.data[:,2:]
plt.scatter(x[y==0,0],x[y==0,1])
plt.scatter(x[y==1,0],x[y==1,1])
plt.scatter(x[y==2,0],x[y==2,1])
```
